```
from speck import SpeckPlot, SpeckWidget
```

```
image_url = 'https://i.imgur.com/JLhMo6E.jpg' #@param {type:'string'}
resize = 60  #@param {type: 'slider', min: 50, max: 250}
upscale = 10 #@param {type: 'slider', min: 5, max: 25}
horizontal = True #@param {type: 'boolean'}

s = SpeckPlot.from_url(image_url, resize=resize, upscale=upscale, horizontal=horizontal)
```

```
SpeckWidget(s).interact()
```
